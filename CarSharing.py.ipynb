```python
# QUESTION 2
# Using appropriate hypothesis testing, determine if there is a significant relationship between 
# each column (except the timestamp column) and the demand rate. Report the tests’ results.

import pandas as pd
from scipy.stats import chi2_contingency
from scipy.stats import pearsonr

# Read the preprocessed CSV file into a DataFrame
df2 = pd.read_csv('PreprocessedCarSharing.csv')

# Perform chi-square test for each categorical column and demand rate
categorical_columns = ['season', 'holiday', 'workingday', 'weather']
for column in categorical_columns:
    contingency_table = pd.crosstab(df2[column], df2['demand'])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    print(f"Chi-square test for {column}: chi2 = {chi2:.4f}, p-value = {p_value:.4f}")

# Perform Pearson correlation test for each continuous column and demand rate
continuous_columns = ['temp', 'temp_feel', 'humidity', 'windspeed']
for column in continuous_columns:
    correlation, p_value = pearsonr(df2[column], df2['demand'])
    print(f"Pearson correlation test for {column}: correlation = {correlation:.4f}, p-value = {p_value:.4f}")
```

```python
# QUESTION 3
# Please describe if you see any seasonal or cyclic pattern in the 
# temp, humidity, windspeed, or demand data in 2017. Describe your answers

import pandas as pd
import matplotlib.pyplot as plt

# Load the preprocessed CSV file
df2 = pd.read_csv('PreprocessedCarSharing.csv')

# Convert the 'timestamp' column to datetime format
df2['timestamp'] = pd.to_datetime(df2['timestamp'])

# Set the 'timestamp' column as the index
df2.set_index('timestamp', inplace=True)

# Extract the variables of interest, excluding columns with object data type
variables = ['temp', 'humidity', 'windspeed', 'demand']

# Resample the data on a monthly basis and compute the mean
monthly_data = df2[variables].resample('M').mean()

# Plot the variables over time (monthly trends)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
monthly_data['temp'].plot(ax=axes[0, 0])
axes[0, 0].set_xlabel('Time')
axes[0, 0].set_ylabel('Temperature')
axes[0, 0].set_title('Monthly Temperature')

monthly_data['humidity'].plot(ax=axes[0, 1])
axes[0, 1].set_xlabel('Time')
axes[0, 1].set_ylabel('Humidity')
axes[0, 1].set_title('Monthly Humidity')

monthly_data['windspeed'].plot(ax=axes[1, 0])
axes[1, 0].set_xlabel('Time')
axes[1, 0].set_ylabel('Windspeed')
axes[1, 0].set_title('Monthly Windspeed')

monthly_data['demand'].plot(ax=axes[1, 1])
axes[1, 1].set_xlabel('Time')
axes[1, 1].set_ylabel('Demand')
axes[1, 1].set_title('Monthly Demand')

plt.suptitle('Monthly Trends')
plt.tight_layout()
plt.show()
```

```python
# QUESTION 4
# Use an ARIMA model to predict the weekly average demand rate. Consider 30 percent of data for testing.

from pandas import read_csv
from datetime import datetime
from matplotlib import pyplot as plt
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
from sklearn.metrics import mean_squared_error

# Read the CSV file with datetime parsing
df2 = pd.read_csv('PreprocessedCarSharing.csv', parse_dates=['timestamp'], index_col='timestamp')

# Calculate weekly average demand rate
weekly_avg_demand = df2['demand'].resample('W').mean().dropna()

# Consider 30 percent of data for testing
train_size = int(len(weekly_avg_demand) * 0.7)
train, test = weekly_avg_demand[:train_size], weekly_avg_demand[train_size:]

# Checking if the time series is stationary
# and set the value of d in the ARIMA model
result = adfuller(weekly_avg_demand)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

# Differencing to stationarize the time series
differenced_series = weekly_avg_demand.diff().dropna()

# Checking if the differenced time series is stationary
result = adfuller(differenced_series)
print('ADF Statistic after differencing: %f' % result[0])    # this set d = 1
print('p-value after differencing: %f' % result[1])

# Setting the value of p in the ARIMA model
plt.rcParams.update({'figure.figsize': (8, 3), 'figure.dpi': 120})
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(weekly_avg_demand.diff().dropna())
axes[0].set_title('1st Order Differencing')
axes[1].set(ylim=(0, 1.2))
plot_pacf(weekly_avg_demand.diff().dropna(), ax=axes[1])
plt.show()

# Setting the value of q in the ARIMA model
plt.rcParams.update({'figure.figsize': (8, 3), 'figure.dpi': 120})
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(weekly_avg_demand.diff().dropna())
axes[0].set_title('Original time series')
axes[1].set(ylim=(0, 1.2))
plot_acf(weekly_avg_demand.diff().dropna(), ax=axes[1])
plt.show()

# Fitting the ARIMA model using the calculated p,d, and q to check the residual Fitting model
model = ARIMA(train, order=(1, 1, 1))  # p=1, d=1, q=1
model_fit = model.fit()
print(model_fit.summary())

# Plot residual errors
residuals = pd.DataFrame(model_fit.resid, columns=['Residuals'])
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()

# Making predictions and printing forecasted and actual weekly average demand rate with corresponding timestamps
history = train.values.tolist()
predictions = []
for t in range(len(test)):
    model = ARIMA(history, order=(1, 1, 1))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    # Print forecasted and actual demand rate with corresponding timestamps
    print("Timestamp:", test.index[t], "Forecasted:", yhat, "Actual:", test.iloc[t])
    history.append(test.iloc[t])

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)

# Plot forecasts against actual outcomes
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, predictions, color='red', label='Predicted')
plt.xlabel('Time')
plt.ylabel('Weekly Average Demand Rate')
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()
```

```python
# QUESTION 5
# Use a random forest regressor and a deep neural network to predict the demand rate and report 
#the minimum square error for each model. Which one is working better? Why? Please describe the reason.

from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Assuming df2 contains the preprocessed data
# One-hot encode categorical variables
df2_encoded = pd.get_dummies(df2, columns=['season', 'holiday', 'workingday', 'weather'])

# Prepare the features (X) and target variable (y) for training
X = df2_encoded.drop('demand', axis=1)
y = df2_encoded['demand']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)

# Predict on test set using Random Forest Regressor
rf_predictions = rf_regressor.predict(X_test)

# Calculate MSE for Random Forest Regressor
rf_mse = mean_squared_error(y_test, rf_predictions)

# Train Deep Neural Network
mlp_regressor = MLPRegressor(random_state=42)
mlp_regressor.fit(X_train, y_train)

# Predict on test set using Deep Neural Network
mlp_predictions = mlp_regressor.predict(X_test)

# Calculate MSE for Deep Neural Network
mlp_mse = mean_squared_error(y_test, mlp_predictions)

# Compare Errors (Mean Squared Errors)
print('Random Forest Regressor MSE:', rf_mse)
print('Deep Neural Network MSE:', mlp_mse)
```

```python
# QUESTION 6
# Categorize the demand rate into the following two groups: 
# demand rates greater than the average demand rate and demand rates less than the average demand rate. 
# Use labels 1 and 2 for the first and the second groups, respectively. 
# Now, use three different classifiers to predict the demand rates’ labels and report the accuracy of all models. 
# Use 30 percent of data for testing. 

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Perform one-hot encoding for categorical variables
df2_encoded = pd.get_dummies(df2, columns=['season', 'holiday', 'workingday', 'weather'])

# Calculate the average demand rate
average_demand_rate = df2_encoded['demand'].mean()

# Create a new column in the DataFrame to store the labels
df2_encoded['label'] = np.where(df2_encoded['demand'] > average_demand_rate, 1, 2)

# Split the data into features (X) and labels (y)
X = df2_encoded.drop(['demand', 'label'], axis=1)
y = df2_encoded['label']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train and evaluate Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
dt_predictions = dt_classifier.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Train and evaluate K-Nearest Neighbors Classifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
knn_predictions = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)

# Train and evaluate Support Vector Machine Classifier
svm_classifier = SVC(random_state=42)
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)

# Models Accuracies Report
print('Decision Tree Classifier Accuracy:', dt_accuracy)
print('K-Nearest Neighbors Classifier Accuracy:', knn_accuracy)
print('Support Vector Machine Classifier Accuracy:', svm_accuracy)
```

```python
# QUESTION 7
# Assume k is the number of clusters. Set k=2, 3, 4, and 12 and use 2 methods to cluster the temp data in 2017. 
# Which k gives the most uniform clusters? 
# (Clusters are called uniform when the number of samples falling into each cluster is close.)


from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Prepare the data
X = df2[['temp']]

# Apply clustering with different values of k using K-means and Agglomerative clustering algorithms
k_values = [2, 3, 4, 12]
results = {}

for k in k_values:
    # K-means clustering
    kmeans = KMeans(n_clusters=k)
    kmeans_labels = kmeans.fit_predict(X)
    kmeans_silhouette = silhouette_score(X, kmeans_labels)

    # Agglomerative clustering
    agglomerative = AgglomerativeClustering(n_clusters=k)
    agglomerative_labels = agglomerative.fit_predict(X)
    agglomerative_silhouette = silhouette_score(X, agglomerative_labels)

    results[k] = {
        'K-means Silhouette': kmeans_silhouette,
        'Agglomerative Silhouette': agglomerative_silhouette
    }

# Determine the most uniform clusters based on silhouette scores
uniform_k = max(results, key=lambda k: min(results[k]['K-means Silhouette'], results[k]['Agglomerative Silhouette']))

# Analyze the results and determine the most uniform clusters
print('Silhouette Scores for Different k Values:')
for k, scores in results.items():
    print(f'k = {k}: K-means Silhouette = {scores["K-means Silhouette"]:.4f}, Agglomerative Silhouette = {scores["Agglomerative Silhouette"]:.4f}')

print(f'\nMost Uniform Clusters: k = {uniform_k}')
```
